```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from pprint import pprint
from newscatcherapi_client import Newscatcher, ApiException

load_dotenv()

client = OpenAI(
    api_key = os.environ.get("OPENAI_API_KEY"),
)

# newscatcher = Newscatcher(
#     api_key= os.environ.get("NEWSCATCHER_API_KEY"), 
# )

def get_chat_completion(prompt, data, model="gpt-3.5-turbo"):
  
  response = client.chat.completions.create(
      model=model,
      messages=[
        {"role": "user", "content": prompt},
        {"role": "system", "content": f"As a Communications Expert, generate questions about concepts Using this {data}."}
      ],
      temperature = 0.7,
      response_format={"type": "json_object"}
    )
  return response.choices[0].message.content


response = get_chat_completion("Tell me information about local policing in the North East of England")
print(response)
```
